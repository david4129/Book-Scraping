## Web Scraping

```python
#Scrape data on books from the website
#Return a DataFrame containing the information

#Returns a list of internal links of books as well as their categories
def return_urls():
  book_url = requests.get('http://books.toscrape.com/catalogue/category/books_1/index.html').text
  list_book = []
  book_parsed = BeautifulSoup(book_url, 'lxml')
  first_body = book_parsed.find_all('li')
  for i in book_parsed.find_all('a', href = re.compile('\.\./books/[A-Za-z0-9]+')):
    new_url = 'http://books.toscrape.com/catalogue/category' + i['href'][2:]
    category = i.text.strip()
    list_book.append([new_url, category])
  return list_book


#Returns a Dataframe consisting of 
def return_books():
  book_dic = {}
  title_list = []
  price_list = []
  ratings = []
  cat_list = []
  packed_url = return_urls()
  for i in packed_url:
    obt_url = requests.get(i[0]).text
    parsed_url = BeautifulSoup(obt_url, 'lxml')
    books = parsed_url.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
    for book in books:
      book_title = book.find('h3').text
      book_price = book.find('p', class_ = 'price_color').text[1:]
      rating = book.article.p['class'][1]
      categories = i[1]

      title_list.append(book_title)
      price_list.append(book_price)
      ratings.append(rating)
      cat_list.append(categories)
  book_dic['Title'] = title_list
  book_dic['Price'] = price_list
  book_dic['ratings'] = ratings
  book_dic['category'] = cat_list
  return pd.DataFrame(book_dic)

return_books()
```
